```python
#完成决策树的练习
import time

from sklearn.datasets import load_iris, fetch_20newsgroups, fetch_california_housing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from sklearn.feature_extraction import DictVectorizer
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np
```

```python
# 加载数据集:对泰坦尼克号进行预测生死
titanic = pd.read_csv('./data/titanic.txt')
titanic.info()# 查看数据集基本信息
```

```python
x = titanic[['pclass', 'age', 'sex']]# 选取特征列，不包括生存状态列
y = titanic['survived']# 选取目标列
print(x.info())  # 用来判断是否有空值
x.describe(include='all')# 查看数据集基本信息
```

```python
mean=x['age'].mean()# 查看平均值
print(mean)
x.loc[:,'age']=x.loc[:,'age'].fillna(mean)# 用平均值填充空值
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=4)# 划分数据集
print(x_train.head())
```

```python
dict = DictVectorizer(sparse=False)# 把数据集变为字典形式，便于后续处理
x_train = dict.fit_transform(x_train.to_dict(orient="records"))
print(type(x_train))
print(dict.get_feature_names_out())# 查看特征名
x_test = dict.transform(x_test.to_dict(orient="records"))# 把测试集也变为字典形式
print(x_train)
```

```python
dec = DecisionTreeClassifier()# 创建决策树模型
dec.fit(x_train, y_train)# 训练模型
print("预测的准确率：", dec.score(x_test, y_test))# 查看准确率
export_graphviz(dec, out_file="tree.dot",
                feature_names=['age', 'pclass=1st', 'pclass=2nd', 'pclass=3rd', 'female', 'male'])# 把决策树导出为dot文件
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=4)# 划分数据集
dict = DictVectorizer(sparse=False)# 把数据集变为字典形式，便于后续处理
x_train = dict.fit_transform(x_train.to_dict(orient="records"))# 把训练集也变为字典形式
x_test = dict.transform(x_test.to_dict(orient="records"))# 把测试集也变为字典形式
dec = DecisionTreeClassifier(max_depth=7,min_impurity_decrease=0.01,min_samples_split=20)# 创建决策树模型，并设置参数
dec.fit(x_train, y_train)# 训练模型
print("预测的准确率：", dec.score(x_test, y_test))# 查看准确率
export_graphviz(dec, out_file="tree1.dot",
                feature_names=dict.get_feature_names_out())# 把决策树导出为dot文件
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=4)# 划分数据集
dict = DictVectorizer(sparse=False)# 把数据集变为字典形式，便于后续处理
x_train = dict.fit_transform(x_train.to_dict(orient="records"))# 把训练集也变为字典形式
x_test = dict.transform(x_test.to_dict(orient="records"))# 把测试集也变为字典形式
rf = RandomForestClassifier(n_jobs=-1)# 创建随机森林模型
param = {"n_estimators": [1500,2000, 5000], "max_depth": [2, 3, 5, 8, 15, 25]}# 设置参数
gc = GridSearchCV(rf, param_grid=param, cv=3)# 创建网格搜索模型
gc.fit(x_train, y_train)# 训练模型
print("准确率：", gc.score(x_test, y_test))
print("查看选择的参数模型：", gc.best_params_)
print("选择最好的模型是：", gc.best_estimator_)
```
