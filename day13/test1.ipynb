```python
# 练习groupby
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
num = pd.read_csv("911.csv")
print(num.info())#查看数据信息
num["timeStamp"] = pd.to_datetime(num["timeStamp"])#将时间戳转换为datetime格式
temp_list = num["title"].str.split(": ").tolist()#将title列中的字符串分割，并存入列表中
cate_list = [i[0] for i in temp_list]#提取分类列表
num["cate"] = pd.DataFrame(np.array(cate_list).reshape((num.shape[0], 1)))#将分类列表赋值给num的cate列
num.set_index("timeStamp", inplace=True)#将时间戳设置为索引
num.head(10)
```

```python
plt.figure(figsize=(20, 6),dpi = 100)#设置画布大小
for group_name,group_data in num.groupby(by = "cate"):
    #绘制折线图
    count_by_month = group_data.resample("MS").count()["title"]#降采样，每个月的求助次数
    print(count_by_month)#打印每个月的求助次数
    x = count_by_month.index#获取时间戳
    print(x)#打印时间戳
    y = count_by_month.values#获取求助次数
    x = [i.strftime("%Y%m%d") for i in x]
    plt.plot(x,y,label=group_name)
plt.xticks(range(len(x)),x,rotation=45)#设置x轴刻度
plt.legend(loc = "best")#显示图例
plt.show()
```

```python
dict_list = {'key1' : ['a', 'b', 'a', 'b',
                      'a', 'b', 'a', 'a'],
            'key2' : ['one', 'one', 'two', 'three',
                      'two', 'two', 'one', 'three'],
            'data1': np.random.randint(1, 10, 8),
            'data2': np.random.randint(1, 10, 8)}#创建字典列表
num_obj = pd.DataFrame(dict_list)#创建DataFrame
print(num_obj)
```

```python
num_obj.info()#查看数据信息
```

```python
k1_sum = num_obj.groupby("key1").mean(numeric_only=True).add_prefix("mean_")#按key1分组后，计算data1，data2的统计信息并附加到原始表格中，并添加表头前缀
print(k1_sum)
```

```python
k1_sum_tf = num_obj.loc[:,['key1','data1', 'data2']].groupby('key1').transform('mean').add_prefix('mean_')#使用transform，分组后计算结果和原本的保持一致
k1_sum_tf
```
